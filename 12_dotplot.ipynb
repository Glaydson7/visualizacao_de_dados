```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("expectativa.csv")
df.sort_values(by="Expectativa de vida", inplace=True)
df
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(df["Expectativa de vida"], df["UF"], zorder=2)
plt.xlabel("Anos")
plt.title("Expectativa de vida no Brasil")
plt.grid(zorder=1)
plt.show()
```
