```python
import matplotlib.pyplot as plt
```

```python
# 1d
plt.figure(figsize=(5, 4))

sequencia = [1, 2, 5, 8, 9]
y = [0, 0, 0, 0, 0]
plt.plot(sequencia, y, "o")
plt.hlines(0, 0, 10)
plt.axis("off")
plt.show()
```

```python
# 3d
plt.figure(figsize=(6, 6))
ax = plt.axes(projection='3d')
x = [4, 8, 2, 3, 5]
y = [1, 2, 5, 9, 1]
z = [3, 3, 8, 2, 7]
ax.scatter3D(x, y, z)
plt.show()
```

```python
import pandas as pd
df = pd.read_csv('paises.csv')
df
```

```python
plt.figure(figsize=(5, 6))
ax = plt.axes(projection='3d')
x = df["Area"]
y = df['Populacao']
z = df['Densidade']
ax.set_xlabel = ("rea")
ax.set_ylabel = ("Populcao")
ax.set_zlabel = ("Densidade")
ax.scatter3D(x, y, z)
plt.show()
```
