```python
import nltk
nltk.download("stopwords")
sw = nltk.corpus.stopwords.words("portuguese")
```

```python
sw.append("nesse")
```

```python
plt.figure(figsize=(6, 6))
wc = wordcloud.WordCloud(height=800, width=800, background_color='white',
                         stopwords=sw)
wc.generate(texto)
plt.axis('off')
plt.imshow(wc)
plt.show()
```
