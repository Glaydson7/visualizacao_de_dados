```python
import matplotlib.pyplot as plt
import pandas as pd
pd.set_option('display.max_columns', 44)
```

```python
df = pd.read_csv('brasil.csv')
df
```

```python
# Cores
plt.figure(figsize=(7, 3))

plt.plot(df['inflacao'], color='blue')
plt.plot(df['pib'], color='red')
plt.title("Dados econômicos")
plt.legend(['Inflalção', 'PIB'])
plt.xticks(range(0, 10), range(2011, 2021))

plt.show()
```

```python
# Estilo
plt.figure(figsize=(5, 4))

plt.plot(df['inflacao'], "--", color='black')
plt.plot(df['pib'], ":", color='black')
plt.title("Dados econômicos")
plt.legend(['Inflalção', 'PIB'])
plt.xticks(range(0, 10), range(2011, 2021))

plt.show()
```

```python
# Espessura
plt.figure(figsize=(5, 4))

plt.plot(df['inflacao'], "--", color='black', linewidth=4)
plt.plot(df['pib'], ":", color='black', linewidth=2)
plt.title("Dados econômicos")
plt.legend(['Inflalção', 'PIB'])
plt.xticks(range(0, 10), range(2011, 2021))

plt.show()
```
