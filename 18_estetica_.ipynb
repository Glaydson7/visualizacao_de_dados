```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme()
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
# sns.relplot(x='sepal_length', y='petal_length', data=df)
# sns.relplot(x='sepal_length', y='petal_length', hue='species', data=df)
# sns.relplot(x='sepal_length', y='petal_length', style='species',s=50 , data=df)
sns.relplot(x='sepal_length', y='petal_length', col='species', data=df)
plt.show()
```
