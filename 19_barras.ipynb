```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme()
```

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
sns.catplot(x='class', y='fare', kind='bar', errorbar=None, data=df)
plt.show()
```

```python
sns.catplot(x='class', y='fare', kind='bar', errorbar=None, hue='sex', data=df)
plt.show()
```

```python
sns.catplot(x='class', y='fare', kind='point',
            errorbar=None, hue='sex', data=df)
plt.show()
```
