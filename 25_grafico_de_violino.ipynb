```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme()
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
# sns.catplot(y='tip', kind='violin', data=df)
# sns.catplot(y='tip', x='day', kind='violin', height=4, data=df)
sns.catplot(y='tip', x='day', kind='violin', hue='sex',
            aspect=2, split=True, height=4, data=df)

plt.show()
```
