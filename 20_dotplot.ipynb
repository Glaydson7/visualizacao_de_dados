```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme()
```

```python
df = sns.load_dataset("car_crashes")
df.head()
```

```python
df.sort_values(by='alcohol', ascending=False, inplace=True)
df.head()
```

```python
sns.catplot(y="abbrev", x="alcohol", height=9, width=9, color="black", data=df)
plt.grid()
plt.show()
```
