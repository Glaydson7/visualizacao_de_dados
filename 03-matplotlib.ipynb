```python
import matplotlib.pyplot as plt
```

```python
corridas = [8.2, 5.9, 6.5, 6.1, 4.8, 4.6, 3.9]
pedaladas = [10.2, 6.5, 7.2, 7.7, 11.6, 5.8, 1.3]
```

```python
plt.plot(corridas, color='red', linewidth=4)
plt.title("Minhas corridas", loc='left', pad=2,
          fontdict={'fontsize': 18, "color": "#0000FF"})
# plt.xlabel("Dias da semana", fontdict={'fontsize': 14, "color": "#9400D3"})
plt.xticks([0, 1, 2, 3, 4, 5, 6], ["dom", "seg",
           "ter", "qua", "qui", "sex", "sab"])
plt.ylabel("Distância percorrida (km)", fontdict={
           'fontsize': 14, "color": "#008000"})
plt.ylim(0, 9)
plt.annotate("Meata semanal atingida", xy=(2, 6.5),
             xytext=(1, 7.5), arrowprops={'arrowstyle': '->'})
plt.show()
```

```python
plt.plot(corridas)
plt.plot(pedaladas)
plt.legend(["Corridas", "Pedaladas"])
plt.title("Meus exercícios", loc='left', pad=12,
          fontdict={'fontsize': 18, "color": "#0000FF"})
plt.ylabel("Distância percorrida (km)", fontdict={
           'fontsize': 14, "color": "#008000"})
plt.xticks([0, 1, 2, 3, 4, 5, 6], ["dom", "seg",
           "ter", "qua", "qui", "sex", "sab"])
plt.show()
```

```python
corridas = [8.2, 5.9, 6.5, 6.1, 4.8, 4.6, 3.9]
pedaladas = [10.2, 6.5, 7.2, 7.7, 11.6, 5.8, 1.3]
plt.title("Meus exercícios na semana", loc="left",
          pad=12, fontdict={'fontsize': 18})
plt.xlabel("Corridas (km)")
plt.ylabel("Pedaladas (km)")
plt.plot(corridas, pedaladas, 'o')
plt.show()
```
