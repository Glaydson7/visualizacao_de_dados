```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
temperaturas = np.random.randint(20, 35, 31)
temperaturas
```

```python
plt.figure(figsize=(7, 4))
plt.plot(temperaturas)
plt.title("Temperatura média - Janeiro", loc="left", fontdict={'fontsize': 18})
plt.xlabel("Dias", fontdict={'fontsize': 14})
plt.ylabel("T(°C)", fontdict={'fontsize': 14})
plt.xticks(range(0, 31, 5), range(1, 32, 5))
plt.grid(axis='both', color='lightgray')
plt.hlines(33, -1, 31, color='red', label="Tempratura crítica")
plt.xlim(-1, 31)
plt.vlines(24, 29, 35, color='orange', label="Chegada de massa de ar quente ")
plt.ylim(19, 35)
plt.fill_between([6, 11], y1=19, y2=35, color='lightgray',
                 label='Período de chuvas')
plt.legend(loc='upper right', bbox_to_anchor=(1.54, 1))
plt.show()
```
