```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme()
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
# sns.catplot(y='tip', x='day', kind='box', data=df)
sns.catplot(y='tip', x='day', kind='box', hue='sex',
            height=5, aspect=1.5, data=df)
plt.show()
```
