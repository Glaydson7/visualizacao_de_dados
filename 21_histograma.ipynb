```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme()
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
sns.displot(x='sepal_length', kind='hist', height=5, bins='fd', data=df)
plt.show()
```

```python
# sns.displot(x='sepal_length', kind='hist', height=5, bins='fd', hue='species', data=df)
# sns.displot(x='sepal_length', kind='hist', height=5, bins='fd', hue='species', element='step', data=df)
# sns.displot(x='sepal_length', kind='hist', height=5, bins='fd', hue='species', element='poly', data=df)
sns.displot(x='sepal_length', kind='hist', height=3,
            bins='fd', col='species', data=df)

plt.show()
```
