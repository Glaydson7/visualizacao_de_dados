### Groupby

```python
df.head()
```

```python
df.loc[:, ["Country", "Rating"]].groupby(
    "Country").mean().sort_values("Rating", ascending=False)
```

```python
df.loc[:, ["Region", "Price"]].groupby("Region").mean().sort_values("Price")
```

```python
df.loc[:, ["Country", "Winery", "Rating"]].groupby(
    ["Country", "Winery"]).mean()
```

## Pivot Table

```python
df.head()
```

```python
df.pivot_table(values="Rating", index="Country", aggfunc="mean").sort_values(
    "Rating", ascending=False).head()
```

```python
df.pivot_table(values="Price", index="Region", aggfunc="mean")
```

```python
df.pivot_table(values="Rating", index="Country",
               columns="Year", aggfunc="mean")
```
