```python
import matplotlib.pyplot as plt
import pandas as pd
pd.set_option('display.max_columns', 44)
```

```python
df = pd.read_csv('iris.csv')
df.head()
```

```python
df['Species'].unique()
```

```python
df.info()
```

```python
plt.plot(df['Petal.Length'], df['Sepal.Length'], "o")
plt.title("Comprimento da pétala vs. Comprimento da sépala")
plt.show()
```

```python
plt.plot(df['Sepal.Width'], df['Petal.Width'], "o")
plt.title("Largura da pétala vs. Largura da sépala")
plt.show()
```

```python
plt.plot(df['Sepal.Length'], df['Petal.Width'], "o")
plt.title("Comprimento da sépala vs. Largura da pétala")
plt.show()
```
