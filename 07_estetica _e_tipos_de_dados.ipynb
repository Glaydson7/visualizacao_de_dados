```python
import matplotlib.pyplot as plt
import pandas as pd
pd.set_option('display.max_columns', 44)
```

```python
df = pd.read_csv('paises.csv')
df
```

```python
plt.figure(figsize=(3, 2))
plt.plot(df['Area'], df['Populacao'], ".")
plt.title("Países da América do Sul")
plt.xlabel("Area")
plt.ylabel("Populacao")
plt.show()
```

```python
df_com_copas = df[df["Copa"] == "S"]
df_com_copas
```

```python
df_com_copa = df[df["Copa"] == "S"]
df_com_copa
```

```python
df_sem_copa = df[df["Copa"] == "N"]
df_sem_copa
```

```python
plt.figure(figsize=(5, 4))
plt.plot(df_com_copa['Area'], df_com_copa['Populacao'],
         "*", color='green', markersize=8)
plt.plot(df_sem_copa['Area'], df_sem_copa['Populacao'],
         ".", color='orange', markersize=8)
plt.title("Países da América do Sul")
plt.xlabel("Area")
plt.ylabel("Populacao")
plt.show()
```
