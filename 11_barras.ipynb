```python
df = pd.read_csv("iris.csv")
df.head()
```

```python
tbl_iris = df.pivot_table(values='Sepal.Length', index='Species', aggfunc='mean').\
    sort_values(by='Sepal.Length', ascending=False).reset_index()

tbl_iris
```

```python
plt.bar(tbl_iris['Species'], tbl_iris['Sepal.Length'])
plt.show()
```
