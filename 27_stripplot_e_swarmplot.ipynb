```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme()
df = sns.load_dataset('tips')
df.head()
```

```python
# sns.catplot(x='day', y='tip', kind='strip', height=4, data=df)
sns.catplot(x='day', y='tip', kind='swarm', aspect=1, height=4, data=df)
plt.show()
```

```python
sns.catplot(x='day', y='tip', kind='box', height=4, data=df)
sns.stripplot(x='day', y='tip', color='black', data=df)
plt.show()
```

```python
sns.catplot(x='day', y='tip', kind='violin', height=4, aspect=1, data=df)
sns.swarmplot(x='day', y='tip', color='black', data=df)
plt.show()
```
